#  CRUD : 간식박스 프로그램

1. 함수 도출 : CRUD 기본
2. 함수 구현
3. 문제점 도출 - 함수 리팩토링
4. 전체흐름 구현 함수

```python
# 간식박스 생성 함수
def make_snackbox():
    snackbox = ['두쫀쿠','홈런볼']
    return snackbox
```

```python
# 간식박스에 추가하는 함수
def add_snackbox(sbox): #간식리스트
    make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    #print(sbox)
    return sbox
```

```python
# 메뉴 출력
def display_menu():
    print("1. 간식추가")
    print("2. 간식박스 확인")
    print("3. 간식 제거")
    print("4. 주문 확정")
```

```python
def order_confirm(sbox):
    make_msg = input('주문을 확정하시겠습니까? (y/n)').lower()
    if make_msg == 'y':
        return True
    else:
        return False
```

```python
# 전체보기 함수
def display_snackbox(sbox):
    print(sbox)
```

```python
# 삭제하는 함수
def del_snackbox(sbox):
    make_msg = int(input("빼고 싶은 간식의 번호를 입력해주세요 : ")) - 1
    del sbox[make_msg]
```

```python
# 수정하는 함수
```

```python
#서비스의 흐름
snack_box = make_snackbox()
print('간식박스가 생성완료', snack_box)
while True:
    display_menu() 
    menu_no = input('메뉴를 선택하세요')
    if menu_no == '1':
        add_snackbox(snack_box) #간식리스트
    elif menu_no == '2' :
        display_snackbox(snack_box)
    elif menu_no == '3' :
        del_snackbox(snack_box)
    elif menu_no == '4' :
        display_snackbox(snack_box)
        result = order_confirm(snack_box) #매개변수 통일
        if result == True:
            break
print('고객님의 주문을 감사드립니다')
```
