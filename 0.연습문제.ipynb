```python
first = float(input('첫번째 숫자를 입력하세요 :'))
second = float(input('두번째 숫자를 입력하세요 :'))
mul = (first*second)
add = (first + second)
output = f'First is {first}, and second is {second}. '\
        + f'So, Multiplication = {mul:.3f},'\
        + f' addition = {add:.2f}'#숫자f의 소수점까지
print(output)
```
