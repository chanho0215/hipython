```python
import random

# "이 동전은 앞면이 나올 확률이 50%야"라고 미리 정함 (Design)
expected_prob = 0.5 
trial_count = 10  # 10번 던지기
heads = 0

print(f"--- 실험 시작 (예상 확률: {expected_prob * 100}%) ---")

for i in range(trial_count):
    # 실제 동전 던지기 (0~1 사이 랜덤 숫자 뽑기)
    if random.random() < expected_prob:
        heads += 1
        result = "앞면"
    else:
        result = "뒷면"
    print(f"{i+1}번째 결과: {result}")

# --- 3. 상대도수 (Relative Frequency): 실제 결과 ---
actual_freq = heads / trial_count

print("-" * 30)
print(f"우리의 예상 (확률): {expected_prob * 100}%")
print(f"실제 결과 (상대도수): {actual_freq * 100}%")
```

```python
exp_p = 0.5 #확률
trial_count = 10 #시행
head_count = 0 #앞면이 나온 건수
#동전의 앞면이 나올 확률 50%
#10번 실행했을 때 상대도수와 확률 비교

for cnt in range(trial_count):
    if random.random() < exp_p :
        head_count += 1
        
print(f'앞면이 나온 횟수 : {head_count} ')
actual_freq = head_count/trial_count

print(f'예상확률 : {exp_p*100}%')
print(f'실제결과(상대도수) : {actual_freq*100}%')
```

```python
total_balls = 7
white_balls = 4
black_balls = 3
picked_cnt = 3 #white2, black1

#전체 경우의 수

total_case = math.comb(total_balls, picked_cnt)
```

```python
#흰공이 2개 나올 경우의 수 (4,2)
white_case = math.comb(white_balls, 2)
#검은공 1개 나올 경우의 수 (3, 1)
black_case = math.comb(black_balls, 1)
```

```python
# 흰공2, 검은공1 경우의 수
event_case = white_case * black_case
```

```python
import random
#시뮬레이션
def run_simulation(trials):
    act_freq = 0
    success_cnt = 0
    sample_space = ['W','W','W','W','B','B','B'] #표본공간
    for i in range(trials):
        drawn_balls = random.sample(sample_space, 3)
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') == 1:
            success_cnt += 1 #흰공2, 검공1개 나온 횟수
    act_freq = success_cnt / trials #상대도수
    return act_freq
```

```python
trials = 10000
simulation_prob = run_simulation(trials)
print(f'시뮬레이션 횟수 {trials}, 결과확률 {simulation_prob*100:.2f}')
```

#확률의 곱셈법칙

```python
#독립 사건
a_prob = 0.5 #주사위 짝수
b_prob = 0.5 #동전 앞면
prob_ind = a_prob * b_prob
print(f'독립사건 확률이 곱 : {prob_ind:.2f}')
```

```python
#종속사건
total = 10
win = 2
prob_A = win/total #A가 당첨될 확률 P(A)
prob_B = (win-1) / (total-1) #A 다음에 B가 당첨될 확률 P(B|A)
prob_AB = prob_A * prob_B #A확률 B확률의 곱 = P(A) * P(B|A)
print(f'종속사건 AB의 확률 : {prob_AB:.2f}')
```

```python
# 전체 표본공간
sample_space = [1, 2, 3, 4, 5, 6]

# 사건 A: 짝수
event_A = [x for x in sample_space if x % 2 == 0]

# 사건 B: 4의 약수
event_B = [x for x in sample_space if 4 % x == 0]

# 교집합 A ∩ B
event_A_and_B = list(set(event_A) & set(event_B))

# 확률 계산
P_A = len(event_A) / len(sample_space)
P_A_and_B = len(event_A_and_B) / len(sample_space)

# 조건부확률 P(B | A)
P_B_given_A = P_A_and_B / P_A

P_B_given_A
```

```python
sample_space = list(range(1,7))
event_A = [x for x in sample_space if x % 2 == 0] #리스트 컴프리헨션 (리스트와 for와 if가 한번에)
event_B = [x for x in sample_space if 4 % x == 0]
event_A_and_B = list(set(event_A) & set(event_B)) #교집합
P_A = len(event_A) / len(sample_space)
P_A_and_B = len(event_A_and_B) / len(sample_space)

P_B_given_A = P_A_and_B / P_A

P_B_given_A
```

```python
x_values = list(range(1,4))
p_x = [1/6, 3/6, 2/6]

exp_value = 0
#기대값
for i in range( len(x_values)):
  exp_value += x_values[i] * p_x[i]

print(f'기대값 : {exp_value:.5f}')
```

```python
# 카드 점수 목록
cards = [10, 20, 50]

# 카드가 뽑힐 확률 (모두 동일)
probability = 1 / len(cards)

# 기댓값 계산
expected_value = 0
for score in cards:
    expected_value += score * probability

print(expected_value)
```
