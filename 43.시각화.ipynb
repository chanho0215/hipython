```python
import folium

# 서울 중심 좌표
seoul_map = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 서울 주요 명소 5곳
places = [
    {"name": "경복궁", "lat": 37.5796, "lon": 126.9770},
    {"name": "남산타워", "lat": 37.5512, "lon": 126.9882},
    {"name": "광화문", "lat": 37.5759, "lon": 126.9768},
    {"name": "동대문 디자인 플라자(DDP)", "lat": 37.5663, "lon": 127.0095},
    {"name": "코엑스", "lat": 37.5125, "lon": 127.0587},
]

# 마커 추가
for place in places:
    folium.Marker(
        location=[place["lat"], place["lon"]],
        popup=place["name"],
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(seoul_map)

# 지도 출력
seoul_map
```

```python
import folium
import requests

home = [37.528449, 127.003901]     # [lat, lon]
aigrows = [37.513166, 126.938541]  # [lat, lon]

# folium 지도
m = folium.Map(location=home, zoom_start=13)

folium.Marker(home, popup="My House", icon=folium.Icon(color="blue")).add_to(m)
folium.Marker(aigrows, popup="아이그로스", icon=folium.Icon(color="red")).add_to(m)

# OSRM은 좌표를 (lon,lat) 순서로 받음!
home_lonlat = (home[1], home[0])
aigrows_lonlat = (aigrows[1], aigrows[0])

url = (
    "https://router.project-osrm.org/route/v1/driving/"
    f"{home_lonlat[0]},{home_lonlat[1]};{aigrows_lonlat[0]},{aigrows_lonlat[1]}"
    "?overview=full&geometries=geojson"
)

res = requests.get(url, timeout=20)
res.raise_for_status()
data = res.json()

# 경로(도로를 따라가는 polyline) 좌표 추출: geojson이라 [lon,lat] 리스트로 옴
route_coords = data["routes"][0]["geometry"]["coordinates"]

# folium은 [lat,lon] 순서로 넣어야 해서 뒤집기
route_latlon = [(lat, lon) for lon, lat in route_coords]

# 경로 그리기
folium.PolyLine(
    route_latlon,
    weight=5,
    opacity=0.85,
    tooltip="도로 경로 (OSRM)"
).add_to(m)

# 화면 자동 맞춤
m.fit_bounds([home, aigrows])

m
```
