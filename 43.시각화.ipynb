```python
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
import matplotlib.pyplot as plt
```

```python
d1 = anscombe[anscombe['dataset']=='I']
d2 = anscombe[anscombe['dataset']=='II']
d3 = anscombe[anscombe['dataset']=='III']
d4 = anscombe[anscombe['dataset']=='IV']
```

```python
d1.describe()
```

```python
d2.describe()
```

```python
plt.plot(d1['x'], d1['y'], 'o')
```

```python
plt.plot(d2['x'], d2['y'], 'o')
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1) #2row, 2col,position
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot( d1['x'], d1['y'] , 'o')
axes2.plot( d2['x'], d2['y'] , 'o')
axes3.plot( d3['x'], d3['y'] , 'o')
axes4.plot( d4['x'], d4['y'] , 'o')

axes1.set_title('dataset I')
axes2.set_title('dataset II')
axes3.set_title('dataset III')
axes4.set_title('dataset IV')

fig.suptitle('Anscombe Dataset 으로 보는 데이터 시각화의 의미')
fig.set_tight_layout(True)
plt.show()
```

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,8)) #차트의 크기
plt.plot(dates, prices, label= '주가', marker = 'o') #x축, y축, 표식
plt.xlabel('날짜')
plt.ylabel('가격')
plt.title('5일간의 주가 변동')
plt.legend()
plt.grid()
plt.ylim(95,110)
plt.xlim(0,4)
```

```python
plt.subplot(1,2,1)
plt.plot(dates, prices, marker= 'o')

plt.subplot(1,2,2)
plt.bar(dates, prices)
plt.savefig('my_chart.png')
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
# 1변량 그래프 : 히스토그램 - 구간, 빈도수
```

```python
plt.hist(data=tips_df, x='total_bill', bins = 10, rwidth=0.8)
```

```python
# total_bill vs. tip
plt.subplot(1,2,1)
plt.hist(data=tips_df, x='total_bill', bins = 10, rwidth=0.8, color='red')


plt.subplot(1,2,2)
plt.hist(data=tips_df, x='tip', bins = 10, rwidth=0.8)
```

```python
# 1변량 그래프 : 박스 플롯
```

```python
plt.boxplot(data=tips_df, x='total_bill')
plt.grid()
plt.show()
```

```python
plt.boxplot([tips_df[tips_df['sex']=='Female']['total_bill'], 
             tips_df[tips_df['sex']=='Male']['total_bill'],
             tips_df[tips_df['sex']=='Female']['tip'],
             tips_df[tips_df['sex']=='Male']['tip']],
            labels=['total_bill/Female','total_bill/Male', 'tip/Female', 'tip/Male'])
plt.grid()
plt.show()
```

```python
tips_df
```

```python
plt.boxplot([tips_df[tips_df['day']=='Thur']['total_bill'],
             tips_df[tips_df['day']=='Fri']['total_bill'],
             tips_df[tips_df['day']=='Sat']['total_bill'],
             tips_df[tips_df['day']=='Sun']['total_bill']],
             labels=['total_bill/Thur','total_bill/Fri','total_bill/Sat', 'total_bill/Sun'])
plt.grid()
plt.title('요일별 계산금액')
plt.ylabel('계산 금액')
plt.xlabel('요일')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.boxplot(
    [tips_df[tips_df['day']=='Thur']['total_bill'],
     tips_df[tips_df['day']=='Fri']['total_bill'],
     tips_df[tips_df['day']=='Sat']['total_bill'],
     tips_df[tips_df['day']=='Sun']['total_bill']],
    
    labels=['목요일', '금요일', '토요일', '일요일'],
    patch_artist=True,                     # 박스 채우기
    boxprops=dict(facecolor='#A7C7E7', alpha=0.7),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='gray'),
    capprops=dict(color='gray'),
    flierprops=dict(marker='o', markerfacecolor='gray', alpha=0.5)
)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.title('요일별 총 결제금액 분포', fontsize=14, pad=10)
plt.ylabel('총 결제금액', fontsize=12)
plt.xlabel('요일', fontsize=12)

plt.show()
```

```python
plt.boxplot([tips_df[tips_df['time']=='Lunch']['total_bill'],
             tips_df[tips_df['time']=='Dinner']['total_bill']],
             labels=['total_bill/Lunch','total_bill/Dinner'])
plt.grid()
plt.show()
```

```python
plt.boxplot([tips_df[tips_df['day']=='Thu']['total_bill']],
             labels=['total_bill/Thu'])
plt.grid()
plt.show()
plt.figure(figsize=(15,15))
```
