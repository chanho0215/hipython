```python
import numpy as np
from scipy import stats

# ==========================================
# 1ë‹¨ê³„: ë°ì´í„° ìˆ˜ì§‘ (ê°€ìƒ ë°ì´í„° ìƒì„±)
# ==========================================
np.random.seed(42)  # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ ì„¤ì •

# ìƒí™©: í†µì¦ ìˆ˜ì¹˜(0~100)ê°€ ì–¼ë§ˆë‚˜ ì¤„ì–´ë“¤ì—ˆëŠ”ê°€? (ìˆ˜ì¹˜ê°€ ë†’ì„ìˆ˜ë¡ íš¨ê³¼ ì¢‹ìŒ)
# ê¸°ì¡´ ì§„í†µì œ(Aê·¸ë£¹): í‰ê·  50ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
group_old = np.random.normal(loc=50, scale=10, size=30)

# ì‹ ê·œ ì§„í†µì œ(Bê·¸ë£¹): í‰ê·  55ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
# (ìš°ë¦¬ëŠ” ì‹ ê·œ ì•½ì´ ë” ì¢‹ë‹¤ê³  ê°€ì •í•˜ê³  ë°ì´í„°ë¥¼ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤)
group_new = np.random.normal(loc=55, scale=10, size=30)

print(f"[ë°ì´í„° í™•ì¸]")
print(f"ê¸°ì¡´ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_old.mean():.2f}")
print(f"ì‹ ê·œ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_new.mean():.2f}")
print("-" * 30)
```

```python
t_stat, p_value = stats.ttest_ind(
    group_new,
    group_old,
    equal_var=False,
    alternative='greater'
)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt # ì‹œê°í™”ë¥¼ ìœ„í•´ ì¶”ê°€

np.random.seed(123)

# [ìˆ˜ì •] í˜„ì‹¤ ë°˜ì˜: í‰ê·  72kg, í‘œì¤€í¸ì°¨ 8kgì¸ ì •ê·œë¶„í¬ ë°ì´í„° ìƒì„±
# (ê· ë“±ë¶„í¬ uniform ëŒ€ì‹  normal ì‚¬ìš©)
weight = np.random.normal(loc=72, scale=8, size=100)

# ê·€ë¬´ê°€ì„¤: í‰ê·  = 70kg
# ëŒ€ë¦½ê°€ì„¤: í‰ê·  != 70kg
test_result = stats.ttest_1samp(weight, 70)

print(f"í‘œë³¸ í‰ê· : {np.mean(weight):.2f} kg")
print(f"P-value: {test_result.pvalue:.4f}")

# í•´ì„ ë¡œì§
if test_result.pvalue < 0.05:
    print("ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ê¸°ê° (ì°¨ì´ê°€ ìˆë‹¤)")
    print("í•´ì„: ì´ ì§‘ë‹¨ì˜ ëª¸ë¬´ê²ŒëŠ” 70kgì´ ì•„ë‹ˆë¼ê³  í†µê³„ì ìœ¼ë¡œ í™•ì‹ í•©ë‹ˆë‹¤.")
else:
    print("ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ì±„íƒ (ì°¨ì´ê°€ ì—†ë‹¤)")
    print("í•´ì„: 70kgê³¼ ë‹¤ë¥´ë‹¤ê³  ë§í•  ì¶©ë¶„í•œ ì¦ê±°ê°€ ì—†ìŠµë‹ˆë‹¤.")
```

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from statsmodels.stats.anova import AnovaRM

np.random.seed(42)

# ============================================================
# ê³µí†µ ìœ í‹¸: ê°€ì •(ë“±ë¶„ì‚°) ì²´í¬ + ê¹”ë” ì¶œë ¥
# ============================================================
def levene_test(df, value_col, group_cols):
    """
    Levene ë“±ë¶„ì‚° ê²€ì •
    group_colsê°€ 1ê°œë©´ one-way, 2ê°œë©´ two-way(ì…€ë³„ ê·¸ë£¹)ë¡œ ë¬¶ì–´ì„œ ê²€ì •
    """
    if len(group_cols) == 1:
        groups = [g[value_col].values for _, g in df.groupby(group_cols[0])]
    else:
        groups = [g[value_col].values for _, g in df.groupby(group_cols)]
    stat, p = stats.levene(*groups)
    return stat, p

def print_header(title):
    print("\n" + "="*70)
    print(title)
    print("="*70)

# ============================================================
# 1) ì¼ì›ë¶„ì‚°ë¶„ì„ (One-way ANOVA) ì‚¬ë¡€
#    ì˜ˆ: ê´‘ê³  A/B/C 3ì¢…ì´ ë§¤ì¶œ(ë˜ëŠ” ì „í™˜) í‰ê· ì— ì°¨ì´ê°€ ìˆëŠ”ê°€?
# ============================================================
print_header("1) ì¼ì›ë¶„ì‚°ë¶„ì„ (One-way ANOVA)")

n = 40
df1 = pd.DataFrame({
    "campaign": np.repeat(["A", "B", "C"], n),
    "sales": np.concatenate([
        np.random.normal(100, 10, n),  # A
        np.random.normal(108, 10, n),  # B (ì¡°ê¸ˆ ë†’ê²Œ)
        np.random.normal(115, 10, n),  # C (ë” ë†’ê²Œ)
    ])
})

# ë“±ë¶„ì‚° ê²€ì •
lev_stat, lev_p = levene_test(df1, "sales", ["campaign"])
print(f"[Levene ë“±ë¶„ì‚° ê²€ì •] stat={lev_stat:.3f}, p={lev_p:.4f}")

# ANOVA
model1 = ols("sales ~ C(campaign)", data=df1).fit()
anova1 = anova_lm(model1, typ=2)
print("\n[ANOVA ê²°ê³¼]")
print(anova1)

# ì‚¬í›„ê²€ì • (Tukey)
print("\n[Tukey HSD ì‚¬í›„ê²€ì •]")
tukey1 = pairwise_tukeyhsd(endog=df1["sales"], groups=df1["campaign"], alpha=0.05)
print(tukey1)
```

```python
# ============================================================
# 2) ì´ì›ë¶„ì‚°ë¶„ì„ (Two-way ANOVA) ì‚¬ë¡€
#    ì˜ˆ: êµìœ¡ë°©ë²•(Method)ê³¼ ë‚œì´ë„(Level)ì´ ì‹œí—˜ì ìˆ˜(score)ì— ë¯¸ì¹˜ëŠ” ì˜í–¥
#    - ì£¼íš¨ê³¼: Method, Level
#    - êµí˜¸ì‘ìš©: Method:Level
# ============================================================
print_header("2) ì´ì›ë¶„ì‚°ë¶„ì„ (Two-way ANOVA)")

n_cell = 25
methods = ["Online", "Offline"]
levels = ["Easy", "Hard"]

rows = []
for m in methods:
    for lv in levels:
        # ê¸°ë³¸ í‰ê· 
        mu = 70
        # method íš¨ê³¼
        if m == "Offline":
            mu += 5
        # level íš¨ê³¼
        if lv == "Hard":
            mu -= 10
        # êµí˜¸ì‘ìš©(ì˜ˆ: Offlineì´ Hardì—ì„œ ë” ì˜ ë²„íŒ€)
        if (m == "Offline") and (lv == "Hard"):
            mu += 4

        scores = np.random.normal(mu, 8, n_cell)
        for s in scores:
            rows.append([m, lv, s])

df2 = pd.DataFrame(rows, columns=["method", "level", "score"])

# ë“±ë¶„ì‚°(ì…€ë³„) ê²€ì •
lev_stat, lev_p = levene_test(df2, "score", ["method", "level"])
print(f"[Levene ë“±ë¶„ì‚° ê²€ì •(ì…€ë³„)] stat={lev_stat:.3f}, p={lev_p:.4f}")

# Two-way ANOVA (êµí˜¸ì‘ìš© í¬í•¨)
model2 = ols("score ~ C(method) * C(level)", data=df2).fit()
anova2 = anova_lm(model2, typ=2)
print("\n[ANOVA ê²°ê³¼ (ì£¼íš¨ê³¼ + êµí˜¸ì‘ìš©)]")
print(anova2)

# (ì„ íƒ) ì…€ í‰ê·  í™•ì¸
print("\n[ì…€ í‰ê· í‘œ]")
print(df2.groupby(["method", "level"])["score"].mean().unstack())
```

```python
# ============================================================
# 3) ë°˜ë³µì¸¡ì • ë¶„ì‚°ë¶„ì„ (Repeated Measures ANOVA) ì‚¬ë¡€
#    ì˜ˆ: ë™ì¼ ì°¸ê°€ìë“¤ì´ 3ê°€ì§€ UI ë²„ì „(A/B/C)ì„ ëª¨ë‘ ì‚¬ìš©í–ˆì„ ë•Œ ì‘ì—…ì‹œê°„(time)ì´ ë‹¤ë¥¸ê°€?
#    - ê°œì¸ì°¨ë¥¼ ì œê±°í•˜ê³  ì¡°ê±´(ë²„ì „) íš¨ê³¼ë¥¼ ê²€ì •
# ============================================================
print_header("3) ë°˜ë³µì¸¡ì • ë¶„ì‚°ë¶„ì„ (Repeated Measures ANOVA)")

n_subjects = 30
subjects = np.arange(1, n_subjects + 1)

# ê°œì¸ë³„ ê¸°ë³¸ ì†ë„(ëœë¤ ì¸í„°ì…‰íŠ¸)
baseline = np.random.normal(60, 6, n_subjects)  # í‰ê·  ì‘ì—…ì‹œê°„ 60ì´ˆ

# ì¡°ê±´ íš¨ê³¼: UI A=ê¸°ë³¸, B=ì•½ê°„ ê°œì„ (-4ì´ˆ), C=ë” ê°œì„ (-8ì´ˆ)
conds = ["A", "B", "C"]
effects = {"A": 0, "B": -4, "C": -8}

rows = []
for i, sub in enumerate(subjects):
    for c in conds:
        time = baseline[i] + effects[c] + np.random.normal(0, 3)  # ì¸¡ì •ì˜¤ì°¨
        rows.append([sub, c, time])

df3 = pd.DataFrame(rows, columns=["subject", "ui", "time"])

# ë°˜ë³µì¸¡ì • ANOVA
aovrm = AnovaRM(df3, depvar="time", subject="subject", within=["ui"])
res3 = aovrm.fit()
print("\n[Repeated Measures ANOVA ê²°ê³¼]")
print(res3)

# (ì„ íƒ) ì¡°ê±´ë³„ í‰ê· 
print("\n[ì¡°ê±´ë³„ í‰ê· ]")
print(df3.groupby("ui")["time"].mean())
```
