# Series

```python
import pandas as pd
```

시리즈 : 컬럼 + 인덱스 형식으로 저장, 관리, (리스트/배열과 유사한 형태)
* 컬럼 : 1차원 데이터의 모음 - 1개 컬럼
* 컬럼들 > 데이터프레임

데이터프레임 : 컬럼들 + 인덱스, 엑셀 표(테이블)과 유사한 형태

```python
s1 =pd.Series([10, 20, 30]) #__init__(self, 리스트객체)
s1[0:2]
```

```python
col_idx = ['A', 'B', 'C']
data = [10, 20, 30]
s2 = pd.Series(data, index=col_idx)
s2 # 판다스의 데이터타입: dtype
```

```python
s2[0] #인덱싱 표식
```

```python
s2['A':'C'] #슬라이싱, 인덱스 이름 기반 추출
```

```python
s2[0:3] #순서기반 인덱스
```

```python
dict_data = {'A': 10, 'B': 20, 'C': 30} #딕셔너리 키 = 시리즈 인덱스 
s3 = pd.Series(dict_data)
```

```python
s3.dtypes #객체의 속성
```

```python
s3.info() #객체의 메소드
```

```python
# 리스트로 시리즈 생성
data = [1, 2, 3, 4, 5]
s1 = pd.Series(data)
print(s1)

# 인덱스 지정하여 시리즈 생성
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
s2 = pd.Series(data, index=index)
print(s2)

# 딕셔너리로 시리즈 생성 - 키가 인덱스가 됨
dict_data = {'a':1, 'b':2, 'c':3, 'd':4}
s3 = pd.Series(dict_data)
print(s3)
```

```python
population = pd.Series([9776, 2729, 1566, 3158], 
                    index=['서울', '부산', '인천', '대구'])
population
```

```python
print(population['서울'])  # 9776 출력
print(population[['서울', '부산']])  # 서울과 부산의 데이터만 선택
population[['서울', '인천']]
population[['대구']]
```

```python
population['제주'] = 400 #새로운 행 추가
```

```python
population
```

```python
# 시리즈 생성
names = pd.Series(['김철수', '이영희', '박민수'])
ages = pd.Series([25, 28, 22])
scores = pd.Series([85, 92, 78])

# 시리즈를 딕셔너리로 묶어 데이터프레임 생성
df = pd.DataFrame({
    '이름': names,
    '나이': ages,
    '점수': scores
})
print(df)
```

```python
data = {
    '이름': ['김철수', '이영희', '박민수'],
    '나이': [25, 28, 22],
    '점수': [85, 92, 78]
}
df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd

# 샘플 데이터프레임 생성
data = {
    '이름': ['김철수', '이영희', '박민수', '정수진', '최영호', '한미래', '윤지원'],
    '나이': [25, 28, 22, 31, 27, 24, 29],
    '점수': [85, 92, 78, 95, 88, 73, 91]
}
df = pd.DataFrame(data)

# 처음 5개 행 보기
print("=== head() ===")
print(df.head())

# 마지막 3개 행 보기
print("\n=== tail(3) ===")
print(df.tail(3))

# 데이터프레임 기본 정보
print("\n=== info() ===")
print(df.info())

# 기술 통계량
print("\n=== describe() ===")
print(df.describe())
```
