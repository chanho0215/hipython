# Series

```python
import pandas as pd
```

```python
s1 =pd.Series([10, 20, 30])
s1[0:2]
```

```python
col_idx = ['A', 'B', 'C']
data = [10, 20, 30]
s2 = pd.Series(data, index=col_idx)
```

```python
s2['A':'C']
```

```python
dict_data = {'A': 10, 'B': 20, 'C': 30}
s3 = pd.Series(dict_data)
```

```python
s3
```

```python
# 리스트로 시리즈 생성
data = [1, 2, 3, 4, 5]
s1 = pd.Series(data)
print(s1)

# 인덱스 지정하여 시리즈 생성
data = [1, 2, 3, 4, 5]
index = ['a', 'b', 'c', 'd', 'e']
s2 = pd.Series(data, index=index)
print(s2)

# 딕셔너리로 시리즈 생성 - 키가 인덱스가 됨
dict_data = {'a':1, 'b':2, 'c':3, 'd':4}
s3 = pd.Series(dict_data)
print(s3)
```

```python
population = pd.Series([9776, 2729, 1566, 3158], 
                    index=['서울', '부산', '인천', '대구'])
print(population['서울'])  # 9776 출력
print(population[['서울', '부산']])  # 서울과 부산의 데이터만 선택
```
