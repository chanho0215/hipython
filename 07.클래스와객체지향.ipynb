```python
class Bicycle:  #python Object 원조부모
  def __init__(self, color, speed):
    self.color = color
    self.sloworfast = speed
  
  
  def move(self):
    print(f'{self.sloworfast} {self.color}색 자전거가 움직입니다.')
    
  
```

```python
b1 = Bicycle('파란', "빠른") #객체생성
b2 = Bicycle('빨간', '느린')
```

```python
b1.move()
```

```python
b2.move()
```
