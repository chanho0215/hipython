## 클래스는 구조
객체를 생성하기 위한 틀이 클래스
- 객체를 구성하는 구성요소들이 있다 : 특성(속성) + 행위(메소드)

장점 - 기본 메소드를 상속받아서 쓸 수 있다. 

```python
class Bicycle:  #python Object 원조부모
  def __init__(self, color, speed):  #생성자
    self.color = color    #속성
    self.sloworfast = speed
  
  
  def move(self):  #함수, 이게 개체화 하면 메소드
    print(f'{self.sloworfast} {self.color}색 자전거가 움직입니다.')
    
  
```

```python
b1 = Bicycle('파란', "빠른") #객체생성
b2 = Bicycle('빨간', '느린')
```

```python
b1.move()
```

```python
b2.move()
```

```python
b3 = Bicycle('노란', '느린')
```

```python
b3.move()
```

# 자동차 클래스 정의

- self.속성명 : 인스턴스 속성
- 속성명 : 클래스 속성 

- __init__(self): 생성자, 매개변수에서 인스턴스 속성 설정
- def 메소드이름(self): 메소드 정의

공통적인 것 (객체로 가지 않는다)
- 클래스 메소드(cls) : 클래스 속성 관리, @classmethod "데코레이터"

- 정적 메소드 : 유틸리티

```python
class Car:
  #클래스의 속성
  instance_cnt = 0
  def __init__(self, color=''): #생성자 정의
    #객체의 속성
    self.color = color
    
    Car.instance_cnt += 1
    print(f'{self.color} 자동차 객체가 생성되었습니다. ({Car.instance_cnt}대)')
    
  def autopilot(self):
    print(f'{self.color} 자동차가 자율주행중입니다.')
    print(f'{Car.instance_cnt}대의 자동차가 운행중입니다.')
    
  @classmethod  
  def count_check(cls):
    print(f'자동차 대수: {cls.instance_cnt}')
  
  @staticmethod
  def type_check(type_code):
    if type_code > 20 :
      print('전기자동차')
    elif type_code < 20 :
      print('내연기관 자동차')
    else :
      print('하이브리드 자동차')
```

# 자동차 객체 생성 

```python
c1 = Car()
c2 = Car('red')
```

```python
c1.autopilot()
```

```python
c2.autopilot()
```

```python
Car.count_check()
```

```python
Car.type_check(25)
```

```python
class Calculator:
  #생성자
  def __init__(self, numbers):
    #인스턴스 속성 정의
    self.numbers = numbers
  
  
  def sum(self):
    total = 0
    for num in self.numbers:
      total += num
    print(total)
    
  def avg(self):
    average = 0
    for num in self.numbers:
      average += num
    average = average / len(self.numbers)
    print(average)
```

```python
cal1 = Calculator([1,2,3,4,5])
```

```python
cal1.sum()
```

```python
cal1.avg()
```

```python
cal2 = Calculator([6,7,8,9,10])
```

```python
cal2.sum()
cal2.avg()
```

Q2. Account 클래스 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다.

은행이름: SC은행 계좌번호: 111-11-111111

```python
import random

class Account:
    def __init__(self, owner, balance):
        self.bank_name = "SC은행"
        self.owner = owner
        self.acc_num = self.create_account_number()
        self.balance = balance

    @staticmethod
    def create_account_number():
        num1 = random.randint(100, 999)
        num2 = random.randint(10, 99)
        num3 = random.randint(100000, 999999)
        return f"{num1}-{num2}-{num3}"

    def info(self):
        print(f"은행이름: {self.bank_name} 계좌번호: {self.acc_num} 잔액: {self.balance}원")
        
    def deposit(self, amount):
        self.balance += amount
        print(f"{amount}원이 입금되었습니다. 현재 잔액은 {self.balance}원입니다.")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("잔액이 부족합니다.")
        else:
            self.balance -= amount
            print(f"{amount}원이 출금되었습니다. 현재 잔액은 {self.balance}원입니다.")
```

```python
Account1 = Account("홍길동", 10000)
```

```python
Account1.info()
```

```python
Account1.deposit(5000)
```

```python
Account1.withdraw(18000)
```

```python
class Animal:
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "멍멍!"

class Cat(Animal):
    def make_sound(self):
        return "야옹!"
```

```python
d1 = Dog()
d1.make_sound()  # "멍멍!"
```

```python
c1 = Cat()
c1.make_sound()  # "야옹!"
```

```python
animals = [d1, c1]
for ani in animals:
    print(ani.make_sound())  # "멍멍!", "야옹!" 출력
```
