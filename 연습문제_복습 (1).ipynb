```python
def chk_dup_num(s) :
    result = True
    val_list = []
    for num in s:
        if num not in val_list: #0123456789
            val_list.append(num)
        else:
            result = False
    if len(val_list) != 10 :
        result = False
    return result
    
print(chk_dup_num('0123456789'))
print(chk_dup_num('012349'))
```
