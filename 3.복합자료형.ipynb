```python
students = [{'num':'1','name':'kim', 'kor':80,'math':90, 'eng' :100},
            {'num':'2','name':'park', 'kor':90, 'math':80, 'eng':90},
            {'num':'3','name':'park', 'kor':90, 'math':80, 'eng':90},
            {'num':'4','name':'park', 'kor':90, 'math':80, 'eng':90},
            {'num':'5','name':'hong', 'kor':75, 'math':60, 'eng':80}] #복합자료형도 넣을 수 있다
```

```python
#인덱싱하기
students[4]['kor']
```

```python
#for문 사용 
for stu in students:
    stu['total'] = stu['kor']+stu['math']+stu['eng'] #total이 생김
    print(stu['total'])
    #print('\n')
```

```python
students
```

```python
#평균 구하기
for stu in students:
    stu['average'] = round((stu['total'] / 3), 2) #round(a,숫자) 는 소수 숫자 자리수로 반올림해준다.
    print(stu['average'])
```

```python
for stu in students:
    print(stu)
```

```python
for stu in students:
    sorted(stu['total'])
```
