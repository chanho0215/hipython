```python
con = pymysql.connect(
  host = 'localhost',
  user = 'root',
  password = '1234',
  db = 'wntrade',
  charset = 'utf8'
  ) 
# 커서 생성
cursor = con.cursor()

#SQL 쿼리

sql = """SELECT 담당자명
, 고객회사명
, 마일리지
, 고객.도시
, 도시_평균마일리지
, 도시_평균마일리지 - 마일리지 AS 차이
FROM 고객
, (SELECT 도시
	, AVG(마일리지) AS 도시_평균마일리지
    FROM 고객
    GROUP BY 도시
) AS 도시별요약
WHERE 고객.도시 = 도시별요약.도시;"""

#커서 실행
cursor.execute(sql)
result = cursor.fetchall()


for row in result:
  print(row)
  
cursor.close()
con.close()
```

```python
#mysql과 연결하는 함수 정의
def getconnection():
  con = pymysql.connect(
  host = 'localhost',
  user = 'root',
  password = '1234',
  db = 'wntrade',
  charset = 'utf8'
  ) 
  return con
```

```python
import pandas as pd
```

```python
sql = """SELECT 담당자명
, 고객회사명
, 마일리지
, 고객.도시
, 도시_평균마일리지
, 도시_평균마일리지 - 마일리지 AS 차이
FROM 고객
, (SELECT 도시
	, AVG(마일리지) AS 도시_평균마일리지
    FROM 고객
    GROUP BY 도시
) AS 도시별요약
WHERE 고객.도시 = 도시별요약.도시;"""
```

```python
#mysql 연결함수 호출
con = getconnection()


try: 
  # 2. 커서 생성 (커서가 유용한 동안)
  with con.cursor() as cursor:
    #3. SQL 실행
    cursor.execute(sql)
    #4. 쿼리 실행 결과 받기
    result = cursor.fetchall()
    # dataframe으로 결과 변환하기
    result_df = pd.DataFrame(result)
  

  
finally:
  # 6. 연결 종료
  con.close()
```
