# 이전대화를 포함하여 답변하기

- system :
- user :
- assistant : 그전에 기억하고 싶은 대화를 저장 -> rag로 간다, 사람이 검수, 중간산출물

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user", "2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role":"assistant",
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user", "그 나라가 화제가 되었던 이유를 설명해줘")
    ]
)
print(response.output_text)
````

````python
def msg_assist(role, text) :
    msg_src = {"role": role, "content": [{"type":"output_text","text":text}]}
    return msg_src
````

````python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user", "사과"),
        msg_assist("assistant","과일" ),
        msg("user", "고양이")
    ]
)
print(response.output_text)
````

# 파라미터의 제어
- temperature : 창의성 0~1 (최대)
- max_output_tokens : 최대 응답 길이
- top_p : 후보선정 기준 설정 0~1 확률값
- frequency_penalty : 반복 억제

````python
top_p = [0.6, 0.8, 1.0]
for p in top_p:
    response = client.responses.create(
        model= "gpt-4o-mini",
        input="마케팅 문구 1개 만들어줘",
        top_p = p
    )
    print(response.output_text)
````

````python
def return_label(text): #text: 고장내용
    response = client.responses.create(
        model="gpt-4o-mini",
        input=[
            msg("system", sys_prompt),
            msg("user", "고장내용: " + text + "\n출력:")
        ],
        temperature=0
    )

    return response.output_text #출력 리턴
````
